### 1. Local Tests

The initial phase involves a series of tests conducted at the level of test nodes. These tests utilize the test loaders from the federated dataset. They are executed on the LearningManager as a concluding step and the resulting data is stored in the database. This stored data is intended for analysis purposes.

##### Global confusion matrix 

The global confusion Matrix is the sum of all global matrixs of all test nodes in the pipeline
to plot the global confusion matrix we use the `plot_global_confusion_matrix` Method that requires the name of the pipeline 

```python
# Plot the Global confusion Matrix of the pipiline "the first fl_pipeline"
AccuracyLossPlotter.plot_global_confusion_matrix(pipeline_name="the second fl_pipeline")
```

##### Confusion matrix by node 

to plot the confusion Matrix by node, we use the mehtode 


```python
# Plot the confusion matrix of the node ''
AccuracyLossPlotter.plot_confusion_Matrix_by_node(node_name="hospital_3" , pipeline_name="the second fl_pipeline")
```

##### Comparision of Classification report between Nodes 

The methode `plot_classification_report` gives a Comparision between the reports of different test nodes participated in one learning pipeline 

```python
# Plot the classification report of nodes for the pipeline 'the first fl_pipeline'
AccuracyLossPlotter.plot_classification_report(pipeline_name="the second fl_pipeline")
```
